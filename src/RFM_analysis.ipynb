```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt, timedelta
import plotly.express as px
import seaborn as sns
import plotly.colors as pc
from ETL_pipeline import file_path
df = pd.read_csv(file_path)
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
sorted_df = df.sort_values(by='Order Date', ascending=True)
sorted_df
```

```python
reference_date = df['Order Date'].max() + timedelta(days=1)
reference_date
```

```python
rfm_df = df.groupby('Customer ID').agg({
    'Order Date' : lambda x: (reference_date - x.max()).days,
    'Order ID' : 'count',
    'Sales' : 'sum'
}).reset_index()
rfm_df.tail()
```

```python
rfm_df_m = rfm_df.rename(columns={'Order Date':'Recency', 'Order ID':'Frequency', 'Sales':'Monetary'})
rfm_df_m
```

```python
Quantiles = rfm_df_m[['Recency', 'Frequency', 'Monetary']].quantile([0.25, 0.5, 0.75, 1])
Quantiles
```

```python
q1_recency = rfm_df_m['Recency'].quantile(0.25)
q2_recency = rfm_df_m['Recency'].quantile(0.5)
q3_recency = rfm_df_m['Recency'].quantile(0.75)

def r_score(x):
    if x <= q1_recency:
        return 4
    elif x > q1_recency and x <= q2_recency:
        return 3
    elif x > q2_recency and x <= q3_recency:
        return 2
    else:
        return 1

rfm_df_m['R'] = rfm_df_m['Recency'].apply(r_score)
```

```python
q1_frequency = rfm_df_m['Frequency'].quantile(0.25)
q2_frequency = rfm_df_m['Frequency'].quantile(0.5)
q3_frequency = rfm_df_m['Frequency'].quantile(0.75)

def f_score(x):
    if x <= q1_frequency:
        return 1
    elif x > q1_frequency and x <= q2_frequency:
        return 2
    elif x > q2_frequency and x <= q3_frequency:
        return 3
    else:
        return 4

rfm_df_m['F'] = rfm_df_m['Frequency'].apply(f_score)
```

```python
q1_monetary = rfm_df_m['Monetary'].quantile(0.25)
q2_monetary = rfm_df_m['Monetary'].quantile(0.5)
q3_monetary = rfm_df_m['Monetary'].quantile(0.75)

def m_score(x):
    if x <= q1_monetary:
        return 1
    elif x > q1_monetary and x <= q2_monetary:
        return 2
    elif x > q2_monetary and x <= q3_monetary:
        return 3
    else:
        return 4
rfm_df_m['M'] = rfm_df_m['Monetary'].apply(m_score)
```

```python
rfm_df_m
```

```python
rfm_df_m['RFM_Segment']=rfm_df_m['R'].astype(str) + rfm_df_m['F'].astype(str) + rfm_df_m['M'].astype(str)
```

```python
rfm_df_m['RFM_Score'] = rfm_df_m['R'] + rfm_df_m['F'] + rfm_df_m['M']
rfm_df_m
```

```python
import numpy as np
def seqmentation(x):
    if x == '444':
        return 'VIP Client'
    elif x in ['344', '434', '443', '433', '343', '334', '333']:
        return 'High-Tier Loyal Customer'
    elif x in ['331', '332', '342', '341', '431', '432', '441', '442']:
        return 'Loyal Customer'
    elif x in ['424', '423', '422', '421', '324', '323', '322', '321']:
        return 'Potential Loyal Customers'
    elif x in ['314', '313', '312', '311', '244', '243', '242', '241', '234', '233', '232', '231', '224', '223', '222', '221']:
        return 'At Risk'
    elif x in ['214', '213', '212', '211', '144', '143', '142', '141',  '134', '133', '132', '131', '124', '123', '122', '121', '114', '113', '112', '111']:
        return 'Lost Client'
    elif x in ['411', '412', '413', '414']:
        return 'New Customer'
        
rfm_df_m['Customer_Category'] = rfm_df_m['RFM_Segment'].apply(seqmentation)
    
```

```python
stat_by_category = rfm_df_m.groupby('Customer_Category').agg(total=('Monetary', 'sum'), median=('Monetary', 'median'))
```

```python
only_total_by_c = stat_by_category[['Customer_Category', 'total']].reset_index()
only_total_by_c
```

```python
fig2 = px.bar(only_total_by_c,
              x='Customer_Category', 
              y='total', 
              title='Total sales for each category',
             labels={'Customer_Category':'Customer Category', 'total':'Total'},
             color='Customer_Category',
             color_discrete_sequence=px.colors.qualitative.Set3)
fig2.update_traces(textfont_size=10)
fig2.update_xaxes(tickangle=330)
fig2.show()
```

```python
only_median_by_c = stat_by_category[['Customer_Category', 'median']].reset_index()
```

```python
fig2 = px.bar(only_median_by_c,
              x='Customer_Category', 
              y='median', 
              title='Median of sales for each category',
             labels={'Customer_Category':'Customer Category', 'median':'Median'},
             color='Customer_Category',
             color_discrete_sequence=px.colors.qualitative.Plotly)
fig2.update_traces(textfont_size=10)
fig2.update_xaxes(tickangle=330)
fig2.show()
```
